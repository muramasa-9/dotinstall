```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.core.common.is_list_like = pd.api.types.is_list_like
```

```python
start = '2019-11-01'
end = '2020-11-01'

df = data.DataReader('^N225', 'yahoo', start, end)
```

```python
df.head(10)
```

```python
df.index
```

```python
date = df.index
price = df['Adj Close']
```

```python
plt.plot(date,price)
```

```python
plt.figure(figsize=(30,10))
plt.plot(date,price,label='Nikkei225')
plt.plot(date,df['sma01'],label='sma01')
plt.plot(date,df['sma02'],label='sma02')
plt.plot(date,df['sma03'],label='sma03')

plt.title('N225',color='white',backgroundcolor='blue',size=30,loc='center')
plt.xlabel('date',color='blue',size=20)
plt.ylabel('price',color='blue',size=20)
plt.legend()
```

```python
span01=5
span02=25
span03=50

df['sma01'] = price.rolling(window=span01).mean()
df['sma02'] = price.rolling(window=span02).mean()
df['sma03'] = price.rolling(window=span03).mean()
```

```python
pd.set_option('display.max_rows', None)
df.head(100)
```

```python
plt.figure(figsize=(30,15))
plt.subplot(2,1,1)
plt.plot(date,price,label='Nikkei2225')
plt.plot(date,df['sma01'],label='sma01')
plt.plot(date,df['sma02'],label='sma02')
plt.plot(date,df['sma03'],label='sma03')
plt.legend()

plt.subplot(2,1,2)
plt.bar(date,df['Volume'],label='Volume', color='red')
plt.legend()
```
